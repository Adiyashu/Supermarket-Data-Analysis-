```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset=pd.read_csv("supermarket_sales - Sheet1.csv")
```

```python
#missing value
dataset.isnull().sum()
#if there is any missing value then fill those miisng value-->>dataset['name of column with misssing value'].fillna(dataset['name of column with misssing value'].mean(),inplace=True)
```

```python
dataset['Customer type'].value_counts()
```

```python
plt.figure(figsize=(4,5))
sns.countplot(x="Customer type",data=dataset)
plt.title("Customer type") 
plt.show()
```

```python
dataset['Product line'].value_counts()
```

```python
x=dataset['Product line'].value_counts().index
dataset['Product line'].value_counts().values
 
```
